Employee functions

---



```python
def sort_books(inventory):
  sorted_books = sorted(inventory.keys())
  print("\nBooks sorted alphabetically:")
  for book in sorted_books:
    print(book)
  print()

# both guest and employee use this function:
def find_book(inventory, book_name):
    # empty dictionary to store matching books:
    matches = {}

    # Loop through each book in the inventory
    for book, details in inventory.items():
        # Check if the entered book_name (in lowercase) is found in any of the dictionary book names (also in lowercase)
        # case insensitive
        if book_name.lower() in book.lower():
            # If it matches, add it to the 'matches' dictionary
            matches[book] = details

    if matches:
        print("\nSearch results:")
        for book, details in matches.items():
            print(f"{book}: ${details['price']} ({details['stock']} in stock)")
    else:
        print(f"No books found with '{book_name}' in the title.")

def remove_book(inventory, book_name):
    book_name_normalized = book_name.strip().title()
    if book_name_normalized in inventory and inventory[book_name_normalized]["stock"] > 0:
        inventory[book_name_normalized]["stock"] -= 1
        print(f"You have removed {book_name_normalized}. New stock: {inventory[book_name_normalized]['stock']}")
    else:
        print(f"Sorry, '{book_name_normalized}' is not available or out of stock.")

def add_to_inventory(inventory, book_name, price, stock, genre):
    book_name_normalized = book_name.strip().title()
    if book_name_normalized in inventory:
        inventory[book_name_normalized]["stock"] += stock
        inventory[book_name_normalized]["price"] = price
        print(f"The stock for '{book_name_normalized}' has been updated. New stock: {inventory[book_name_normalized]['stock']}")
    else:
        inventory[book_name_normalized] = {
            "price": price,
            "stock": stock,
            "genre": genre
        }
        print(f"'{book_name_normalized}' has been added to the inventory.")



def find_total_income():
  print(f"The store's current income is {total_income}.")

def find_total_books():
  total_books = sum(book["stock"] for book in inventory.values())
  print(f"The store's current stock is {total_books}.")

def change_book_price():
 book_name = input("Enter the name of the book you want to update: ")
 book_name_normalized = book_name.strip().title()
 if book_name_normalized in inventory:
    new_price = float(input(f"Enter the new price for '{book_name_normalized}': "))
    inventory[book_name_normalized]["price"] = new_price
    print(f"The price for '{book_name_normalized}' has been updated to ${new_price:.2f}.")
 else:
  print(f"'{book_name_normalized}' is not found in the inventory so its price cannot be updated.")
```

Guest functions

---



```python
requested_books = []
import random # module

def request_book(requested_books, book_name):
    requested_books.append(book_name)
    print(f"{book_name} has been added to the request list.")
    print(f"All book requests: {', '.join(requested_books)}")

total_income = 0.0

def pay_for_book(inventory, book_name):
    global total_income
    book_name_normalized = book_name.strip().title()
    if book_name_normalized in inventory and inventory[book_name_normalized]["stock"] > 0:
        print(f"You have purchased '{book_name_normalized}' for ${inventory[book_name_normalized]['price']:.2f}.")
        total_income += inventory[book_name_normalized]["price"]
        inventory[book_name_normalized]["stock"] -= 1
    else:
        print(f"Sorry, '{book_name_normalized}' is out of stock or not available.")

def steal_book(inventory, book_name):
    book_name_normalized = book_name.strip().title()
    if book_name_normalized in inventory and inventory[book_name_normalized]["stock"] > 0:
        inventory[book_name_normalized]["stock"] -= 1
        print(f"You have stolen '{book_name_normalized}'! There are {inventory[book_name_normalized]['stock']} left.")
    else:
        print(f"Sorry, '{book_name_normalized}' is out of stock or not available.")

def recommend_book_by_genre(inventory, genre):
    matching_books = [book for book, details in inventory.items() if genre.lower() in details["genre"].lower()]
    if len(matching_books) > 0:
        recommended_book = random.choice(matching_books)
        book_details = inventory[recommended_book]
        return f"We recommend '{recommended_book}'! It costs ${book_details['price']} and is in stock ({book_details['stock']} available)."
    else:
        return f"No books found in the genre '{genre}'. Please try another genre."
```

Main program

```python
# Welcome
print("Welcome to our bookstore, The Python Page!")
user_type = input("Are you a guest or employee? (guest/employee):").lower()

# Employee choice of actions
if user_type == "employee":
  while True:
    print("What do you want to do? (type an integer from 1-8)")
    print(
        "1. Sort books\n2. Find a book\n3. Add a book\n4. Remove a book\n"
        "5. View the store's income\n6. how many books do we have\n7. adjust prices for books\n8. Leave store"
    )
    choice = input("Enter your choice: ")

    if choice == "1":
      sort_books(inventory)
    elif choice == "2":
      book_name = input("Enter part of the book name to search: ")
      find_book(inventory, book_name)

    elif choice == "3":
      book_name = input("Enter the name of the book: ").strip()
      price = float(input("Enter the price of the book: $").strip())
      stock = int(input("Enter the stock quantity: ").strip())
      genre = input("Enter the genre of the book: ").strip()
      add_to_inventory(inventory, book_name, price, stock, genre)

    elif choice == "4":
      book_name = input("What is the title of the book you'd like to remove? ")
      remove_book(inventory, book_name)
    elif choice == "5":
      find_total_income()
    elif choice == "6":
      find_total_books()
    elif choice == "7":
      change_book_price()
    elif choice == "8":
      print("See you next time!")
      break
    else:
      print("Invalid choice. Please enter an integer from 1-8.")

# Guest choice of actions
elif user_type == "guest":
  while True:
    print("\nWhat do you want to do?")
    print(
        "1. Continue walking\n2. Find a book\n3. Request a book\n4. Pay for a book\n"
        "5. Steal a book\n6. Be reccomended a book\n7. Leave store"
    )
    print()
    choice = input("Enter your choice: ")

    if choice == "1":
      print("Enjoy browsing!")

    elif choice == "2":
      book_name = input("Enter part of the book name to search: ")
      find_book(inventory, book_name)
      print()
    elif choice == "3":
      book_name = input("What is the title of the book you'd like to request? ")
      request_book(requested_books, book_name)

    elif choice == "4":
        book_name = input("What is the title of the book you'd like to pay for? ")
        pay_for_book(inventory, book_name)

    elif choice == "5":
      book_name = input("What is the title of the book you'd like to steal? ")
      steal_book(inventory, book_name)

    elif choice == "6":
      genre= input("What's your favorite genre?")
      print(recommend_book_by_genre(inventory, genre))

    elif choice == "7":
      print("See you next time!")
      break
    else:
      print("Invalid choice. Please enter an integer from 1-7.")


else:
  print('Invalid user type. Please enter "employee" or "guest". Try again.')
```
